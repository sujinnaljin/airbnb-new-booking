```
df_train = pd.read_csv('/content/drive/My Drive/airbnb-recruiting-new-user-bookings/processed_train_users.csv')
df_test = pd.read_csv('/content/drive/My Drive/airbnb-recruiting-new-user-bookings/processed_test_users.csv')

df_all=pd.concat((df_train, df_test), axis = 0, ignore_index = True)
```

```
#One-hot-encoding features
ohe_feats = ['gender', 'signup_method', 'signup_flow', 'language', 'affiliate_channel', 'affiliate_provider', 'first_affiliate_tracked', 'signup_app', 'first_device_type', 'first_browser',  'dac_weekday',  'dac_season', 'tfa_season', 'tfa_weekday', 'tfa_trange', 'dac_year', 'tfa_year']
for f in ohe_feats:
    df_all_dummy = pd.get_dummies(df_all[f], prefix=f)
    df_all = df_all.drop([f], axis=1)
    df_all = pd.concat((df_all, df_all_dummy), axis=1)
```

```
# normalize
from sklearn import preprocessing
standardScaler = preprocessing.StandardScaler()

z_Score = ['age', 'dac_pred_diff', 'tfa_pred_diff', 'dac_month', 'dac_day', 'tfa_month', 'tfa_day', 'tfa_time']
for f in z_Score:
    df = pd.DataFrame(df_train[f])
    all_=pd.DataFrame(df_all[f])
    x = df[[f]].values.astype(float)
    all_x=all_[[f]].values.astype(float)
    standardScaler.fit(x) # train set의 mean, std 이용
    all_x_scaled =standardScaler.transform(all_x)
    df_all_normalized = pd.DataFrame(all_x_scaled)
    df_all[f] = df_all_normalized

df_all.head()
```
